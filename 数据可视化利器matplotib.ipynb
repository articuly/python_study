```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
ds=pd.read_csv('.\\data_set\\pop.csv',skiprows=4)
ds.head(5)
```

```python
ds.drop(['Country Code','Indicator Name','Indicator Code','1960','2016','Unnamed: 61'],axis=1,inplace=True)
ds.dropna(axis=0,how='any',inplace=True)
ds.isnull().sum()
```

```python
x=ds.columns[1:]
y=ds.iloc[37][1:]
country=ds.iloc[37][0]
plt.plot(x,y)
```

```python
plt.rcParams['figure.figsize']=(30,30)
plt.rcParams['font.size']='20'
plt.title('Population density of '+country+' over the years')
plt.xlabel('years')
plt.xticks(rotation='90')
plt.ylabel('Population density (People per sq. Km')
plt.plot(x,y,linewidth=5)
```

```python
x=ds.columns[1:]
colors=cm.rainbow(np.linspace(0,1,5))
for i in range(5):
    y=ds.iloc[i][1:]
    plt.plot(x,
             y,
             c=colors[i],
             label=ds.iloc[i][0],
             linewidth=5)
    plt.title('Comparing population density of various countries')
    plt.xlabel('Years')
    plt.xticks(rotation='90')
    plt.ylabel('Population Density')
    plt.legend(prop={'size':30})
```

```python
countries=ds['Country Name']
populationdensity2015=ds['2015']
plt.xticks(rotation='90')
plt.bar(countries,populationdensity2015,color=cm.rainbow(np.linspace(0,1,len(countries))))
```

```python
top20=ds.sort_values('2015',ascending=False).head(20)
plt.xticks(rotation='45')
plt.title('Population Density of 20 most densely populated countries for the year 2015')
plt.xlabel('Countries')
plt.ylabel('Population Density')
plt.bar(top20['Country Name'],top20['2015'],color=cm.rainbow(np.linspace(0,1,len(top20))))
```

```python
total_columns = ds.shape[1]
selected_data = ds[ds.sum(axis = 1).apply(lambda x: x/total_columns) <= 10]
consolidated_data = selected_data.sum(axis = 1).apply(lambda x: x/total_columns)
countries = selected_data['Country Name']
plt.rcParams['figure.figsize'] = (20, 20)
plt.rcParams['font.size'] = 14
plt.title('Average Population density for various countries')
plt.xlabel('Countries')
plt.ylabel('Average Population Density')
plt.xticks(rotation = '90')
plt.scatter(countries, 
            consolidated_data, 
            s = consolidated_data*30, 
            c = cm.rainbow(np.linspace(0, 1, len(countries))))
```

```python
minimum = ds.loc[:, ds.columns != 'Country Name'].min()
maximum = ds.loc[:, ds.columns != 'Country Name'].max()
diff = maximum - minimum
minOfMax = maximum.min()
plt.title('Range of Population Density for years 1962-2015')
plt.xticks(rotation = '90')
plt.xlabel('Years')
plt.ylabel('Population Density')
plt.bar(ds.columns[1:], diff.apply(lambda x: x-minOfMax), color = cm.rainbow(np.linspace(0, 1, ds.shape[1])))
```
